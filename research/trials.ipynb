```python
#now about the ensure annotations some times we may put the wrong data type in the function so ensure annotations avoid that happening

def get_product(x: int , y: int) -> int:
    return x*y

get_product(3,4)



```

```python
#if i put the wrong data type
get_product(3,'4')
```

```python
from ensure import ensure_annotations

@ensure_annotations
def get_product(x: int , y: int) -> int:
    return x*y

get_product(3,'4')
```
